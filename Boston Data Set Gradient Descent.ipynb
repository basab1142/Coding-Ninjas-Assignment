```python
points = np.array(df.drop("13",axis=1))
```

```python
points.shape
```

```python
def step_gradient(points, learning_rate, m,c):
    m_slope = np.zeros(13)
    c_slope = 0
    M=len(points)
    for i in range(len(points)):
        x = points[i]
        y = np.array(df["13"])[i]
        for j in range(13):
            m_slope[j] = m_slope[j]-(2/M)* (y - (m * x).sum() )*x[j]
            
            
    for j in range(13):
        m[j] = m[j] - learning_rate*m_slope[j]
        c_slope = c_slope + (-2/M)* (y - (m * x).sum() )
    c = c -learning_rate*c_slope
    return m,c
```

```python
def gradient_descent(points, learning_rate, num_iterations):
    m=np.zeros(13)
    c = 0
    for i in range(num_iterations):
        m ,c= step_gradient(points, learning_rate, m,c)
    return m,c
```

```python
def costfunction(points, m):
    cost = 0
    M = len(points)
    for i in range(M):
        x = points[i]
        y = df["13"][i]
        cost += (1/M)*((y - (m*x).sum() )**2)
    return cost
```

```python
def run():
    learning_rate = 0.001
    num_iterations = 1000
    m,c = gradient_descent(points, learning_rate, num_iterations)
    return m,c
```

```python
m,c=run()
```
